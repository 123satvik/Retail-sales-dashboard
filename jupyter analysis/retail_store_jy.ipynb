```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("clean_retail_sales.csv")

df.head()
```

```python
df["total_sales"] = df["RETAIL SALES"] + df["WAREHOUSE SALES"]
```

```python
sales_type = df.groupby("ITEM TYPE")["total_sales"].sum().sort_values(ascending=False)

sales_type.plot(kind="bar")
plt.title("Sales by Item Type")
plt.show()
```

```python
monthly_sales = df.groupby("MONTH ")["total_sales"].sum()

monthly_sales.plot(kind="line", marker="o")
plt.title("Monthly Sales Trend")
plt.show()
```

```python
top_suppliers = df.groupby("SUPPLIER")["total_sales"].sum().sort_values(ascending=False).head(10)

top_suppliers.plot(kind="bar")
plt.title("Top 10 Suppliers")
plt.show()
```
